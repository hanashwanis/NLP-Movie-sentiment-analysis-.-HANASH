```python
import os
import cv2
import numpy as np
from torch.utils.data import Dataset

class UnderwaterDataset(Dataset):

    def __init__(self, image_dir, mask_dir, transform=None):
        self.image_dir = image_dir
        self.mask_dir = mask_dir
        self.transform = transform

        self.images = sorted(os.listdir(image_dir))
        self.masks = sorted(os.listdir(mask_dir))

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):

        image_path = os.path.join(self.image_dir, self.images[idx])
        mask_path = os.path.join(self.mask_dir, self.masks[idx])

        image = cv2.imread(image_path)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        mask = cv2.imread(mask_path, 0)

        if self.transform:
            augmented = self.transform(
                image=image,
                mask=mask
            )

            image = augmented["image"]
            mask = augmented["mask"]

        return image, mask.long()
```

```python
import albumentations as A
from albumentations.pytorch import ToTensorV2

transform = A.Compose([
    A.Resize(256, 256),
    A.Normalize(),
    ToTensorV2()
])
```

```python
DATASET = "/content/dataset/UNDER WATER IMAGE SEGMENTATION"

train_dataset = UnderwaterDataset(
    DATASET + "/images",
    DATASET + "/masks",
    transform=transform
)

val_dataset = UnderwaterDataset(
    DATASET + "/val_images",
    DATASET + "/val_masks",
    transform=transform
)

print(len(train_dataset))
print(len(val_dataset))
```

```python
from torch.utils.data import DataLoader

train_loader = DataLoader(
    train_dataset,
    batch_size=8,
    shuffle=True,
    num_workers=2
)

val_loader = DataLoader(
    val_dataset,
    batch_size=8,
    shuffle=False,
    num_workers=2
)

print(len(train_loader))
print(len(val_loader))
```

```python
images, masks = next(iter(train_loader))

print(images.shape)
print(masks.shape)
print(np.unique(masks.numpy()))
```

```python
import torch

device = torch.device(
    "cuda" if torch.cuda.is_available() else "cpu"
)

print(device)
print(torch.cuda.get_device_name(0))
```

```python
from torchvision import models
import torch.nn as nn

class MobileUNet(nn.Module):

    def __init__(self, num_classes=4):
        super().__init__()

        backbone = models.mobilenet_v2(
            weights=models.MobileNet_V2_Weights.DEFAULT
        )

        self.encoder = backbone.features

        self.decoder = nn.Sequential(

            nn.ConvTranspose2d(1280, 512, 2, stride=2),
            nn.ReLU(),

            nn.ConvTranspose2d(512, 256, 2, stride=2),
            nn.ReLU(),

            nn.ConvTranspose2d(256, 128, 2, stride=2),
            nn.ReLU(),

            nn.ConvTranspose2d(128, 64, 2, stride=2),
            nn.ReLU(),

            nn.ConvTranspose2d(64, 32, 2, stride=2),
            nn.ReLU(),

            nn.Conv2d(32, num_classes, kernel_size=1)
        )

    def forward(self, x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x


model_mobile = MobileUNet(num_classes=4).to(device)

print(model_mobile)
```

```python
total_params = sum(
    p.numel() for p in model_mobile.parameters()
)

print("Total Parameters:", total_params)
```

```python
criterion = nn.CrossEntropyLoss()

optimizer = torch.optim.Adam(
    model_mobile.parameters(),
    lr=0.001
)

mobile_losses = []
```

```python
from tqdm import tqdm

epochs = 15

model_mobile.train()

for epoch in range(epochs):

    running_loss = 0

    loop = tqdm(train_loader)

    for images, masks in loop:

        images = images.to(device)
        masks = masks.to(device)

        optimizer.zero_grad()

        outputs = model_mobile(images)

        loss = criterion(outputs, masks)

        loss.backward()

        optimizer.step()

        running_loss += loss.item()

        loop.set_description(
            f"Epoch [{epoch+1}/{epochs}]"
        )

        loop.set_postfix(
            loss=loss.item()
        )

    epoch_loss = running_loss / len(train_loader)

    mobile_losses.append(epoch_loss)

    print(
        f"Epoch {epoch+1} Loss: {epoch_loss:.4f}"
    )
```

```python
print(mobile_losses)
```

```python
model_mobile.eval()

correct = 0
total = 0

with torch.no_grad():

    for images, masks in val_loader:

        images = images.to(device)
        masks = masks.to(device)

        outputs = model_mobile(images)

        preds = torch.argmax(outputs, dim=1)

        correct += (preds == masks).sum().item()
        total += masks.numel()

accuracy = correct / total

print("Pixel Accuracy:", round(accuracy,4))
```

```python
intersection = 0
union = 0

model_mobile.eval()

with torch.no_grad():

    for images, masks in val_loader:

        images = images.to(device)
        masks = masks.to(device)

        outputs = model_mobile(images)

        preds = torch.argmax(outputs, dim=1)

        intersection += ((preds > 0) & (masks > 0)).sum().item()

        union += ((preds > 0) | (masks > 0)).sum().item()

iou = intersection / union

print("Mean IoU:", round(iou,4))
```

```python
intersection = 0
pred_sum = 0
mask_sum = 0

model_mobile.eval()

with torch.no_grad():

    for images, masks in val_loader:

        images = images.to(device)
        masks = masks.to(device)

        outputs = model_mobile(images)

        preds = torch.argmax(outputs, dim=1)

        intersection += ((preds > 0) & (masks > 0)).sum().item()

        pred_sum += (preds > 0).sum().item()

        mask_sum += (masks > 0).sum().item()

dice = (2 * intersection) / (pred_sum + mask_sum)

print("Dice Score:", round(dice,4))
```

```python
import timm
import torch.nn as nn

class GhostSegNet(nn.Module):

    def __init__(self, num_classes=4):
        super().__init__()

        self.encoder = timm.create_model(
            "ghostnet_100",
            pretrained=True,
            features_only=True
        )

        encoder_channels = self.encoder.feature_info.channels()

        self.decoder = nn.Sequential(

    nn.ConvTranspose2d(160, 256, 2, stride=2),
    nn.ReLU(),

    nn.ConvTranspose2d(256, 128, 2, stride=2),
    nn.ReLU(),

    nn.ConvTranspose2d(128, 64, 2, stride=2),
    nn.ReLU(),

    nn.ConvTranspose2d(64, 32, 2, stride=2),
    nn.ReLU(),

    nn.ConvTranspose2d(32, 16, 2, stride=2),
    nn.ReLU(),

    nn.Conv2d(16, num_classes, kernel_size=1)
)
    def forward(self, x):

        features = self.encoder(x)

        x = features[-1]

        x = self.decoder(x)

        return x
```
